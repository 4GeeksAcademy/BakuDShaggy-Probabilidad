## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import random
import numpy as np
def tirar_el_dado(n_simulaciones = 1000):
    count = 0

    for i in range(n_simulaciones):
        dado1= random.randint(1,6)
        dado2= random.randint(1,6)
        resultado = dado1 + dado2

        if (resultado % 2 ==0) or (resultado > 7):
            count += 1

    return count / n_simulaciones

print(f"La porbabilidad de tirar el dado y que salga un numero par o mayor que 7 es:" , {np.round(tirar_el_dado() * 100 , 2)} , "%")

```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import numpy as np

ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)

# TODO

def tomar_bolas (n_experimentos = 1000):
    cuenta_1 = 0
    cuenta_2 = 0
    for i in range(n_experimentos):
        colors = []

        for i in range (5):
            colors.append(ball_box[np.random.randint(0, 59)])
        colors = np.array(colors)  

        bolas_blancas = sum(colors == "White")
        bolas_verdes = sum(colors == "Green")
        bolas_rojas = sum(colors == "Red")

        if (bolas_blancas == 3) and (bolas_rojas == 2):
            cuenta_1 += 1 
        if (bolas_blancas == 5) or (bolas_rojas == 5) or (bolas_verdes == 5):
            cuenta_2 += 1

    return cuenta_1 / n_experimentos , cuenta_2 / n_experimentos

probabilidad = tomar_bolas(1000)

print(f"La probabilidad que salgan 3 blancas y 2 rojas es: {np.round(probabilidad[0] * 100, 2)}%")
print(f"La probabilidad que salgan todas del mismo color es: {np.round(probabilidad[1] * 100, 2)}%")
```
