## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np

def probabilidad(simulaciones):

    conteo = 0

    for i in range(simulaciones):
        dado_1 = np.random.randint(1,6)
        dado_2 = np.random.randint(1,6)

        suma = dado_1 + dado_2

        if (suma%2 == 0) or (suma > 7):
            conteo += 1
    
    return conteo/simulaciones

print(f'La probabilidad de que los dados sean par o mayores que 7 es del: {np.round(probabilidad(1000) * 100, 2)} %')





```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Creamos el diccionario coon los valores de la cantidad de bolas y sus colores
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)

def probabilidad_2(simulaciones_2): #creamos la funcion 

    tres_y_dos = 0
    iguales= 0

    for i in range(simulaciones_2): #bucle el numero de veces que indicamos en la funcion simulaciones

        tiradas = [] #creamos un array vacio

        for i in range(5): #Creamos un bucle del numero de bolas que vamos a coger
            tiradas.append(ball_box[np.random.randint(0,59)]) #Añadimos al array los valores de la bolsa de bolas aleatorio

        tiradas = np.array(tiradas) #Convertimos con numpy el array tiradas para que no de error
        
        blancas = sum(tiradas == 'White') #sumamos las bolas que han salido en el bucle 5
        rojas = sum(tiradas == 'Red')
        verdes = sum(tiradas == 'Green')

        if (blancas == 3) and (rojas == 2): #comparamos los valores y si es positivo añadimos uno a los contadores
            tres_y_dos += 1
        
        elif (blancas == 5) or (rojas == 5) or (verdes == 5):
            iguales += 1

    return tres_y_dos / simulaciones_2, iguales / simulaciones_2 #sacamos 2 resultados

print(f'La probabilida de que salgan 3 blancas y 2 rojas es de {np.round(probabilidad_2(1000)[0] * 100,2)} %') #printamos los resultados con [0] y [1] de los 2 valores que nos ha devuelto
print(f'La probabilidad de que las 5 sean iguales es del {np.around(probabilidad_2(1000)[1] * 100,2)} %')
    

    
```
